1. What is NumPy, and why is it widely used?
NumPy is a Python library for numerical computing. It is widely used because it provides efficient, fast handling of large multidimensional arrays and matrices compared to standard Python lists.
2. How does broadcasting work in NumPy?
Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by virtually "stretching" the smaller array to match the dimensions of the larger one.
3. What is a Pandas DataFrame?
A DataFrame is a 2-dimensional, labeled data structure with rows and columns, similar to an Excel spreadsheet or SQL table.
4. Explain the use of the groupby() method.
groupby() is used to split data into groups based on some criteria, apply a function (like sum or mean) to each group, and combine the results.
5. Why is Seaborn preferred for statistical visualizations?
Seaborn is built on Matplotlib but offers a higher-level interface, attractive default themes, and specialized functions for statistical plots (like regression and distributions) with less code.
6. Differences between NumPy arrays and Python lists?
NumPy arrays are fixed-size, homogeneous (same data type), and support vectorized operations, making them faster and more memory-efficient than dynamic, heterogeneous Python lists.
7. What is a heatmap, and when should it be used?
A heatmap is a graphical representation of data where values are depicted by color. It is best used to visualize correlation matrices or the intensity of variables across two dimensions.
8. What does "vectorized operation" mean?
It means applying an operation to an entire array at once without writing an explicit loop, leveraging low-level optimizations for speed.
9. How does Matplotlib differ from Plotly?
Matplotlib creates static, publication-quality plots. Plotly creates interactive, web-based plots allowing zooming, hovering, and dynamic resizing.
10. Significance of hierarchical indexing in Pandas?
It allows you to have multiple index levels (MultiIndex) on an axis, enabling you to work with higher-dimensional data in a lower-dimensional (2D) data structure.
11. Role of Seabornâ€™s pairplot() function?
pairplot() creates a grid of plots showing the pairwise relationship between all numerical variables in a DataFrame (scatter plots for joint relationships, histograms for distributions).
12. Purpose of the describe() function?
It generates descriptive statistics for a DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartiles.
13. Why is handling missing data important?
Missing data (NaN) can cause errors in calculations and bias in machine learning models. Handling it ensures data integrity and accurate analysis.
14. Benefits of using Plotly?
Interactive charts (zoom/pan), easy integration with web apps (Dash), support for 3D plots, and highly aesthetic defaults.
15. How does NumPy handle multidimensional arrays?
It uses the ndarray object, which stores data in a contiguous block of memory and uses shape and strides to interpret the dimensions.
16. Role of Bokeh in data visualization?
Bokeh is designed for creating interactive, scalable visualizations for modern web browsers, often used for dashboarding.
17. Difference between apply() and map() in Pandas?
apply() works on entire DataFrames, rows, or columns. map() is used strictly on a Series to substitute each value with another value.
18. Advanced features of NumPy?
Broadcasting, Linear Algebra operations (numpy.linalg), Fourier Transforms, and random number generation.
19. How does Pandas simplify time series analysis?
It provides built-in tools for date range generation, resampling (changing frequency), window functions (rolling stats), and time-zone handling.
20. Role of a pivot table in Pandas?
It summarizes data by grouping and aggregating values, allowing you to reshape data (similar to Excel PivotTables) to analyze relationships between variables.
21. Why is NumPy slicing faster than Python list slicing?
NumPy slicing creates a "view" of the original array (referencing the same memory), whereas Python lists create a completely new copy of the data.
22. Common use cases for Seaborn?
Visualizing correlation matrices (heatmaps), comparing distributions (box/violin plots), and visualizing regression models (lmplot).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os

print("--- STARTING PRACTICAL SOLUTIONS ---\n")

# ==========================================
# 1. Create a 2D NumPy array and calculate the sum of each row
# ==========================================
print("--- Q1: NumPy Row Sum ---")
arr_2d = np.array([[10, 20, 30],
                   [40, 50, 60],
                   [70, 80, 90]])
row_sums = np.sum(arr_2d, axis=1)
print(f"Original Array:\n{arr_2d}")
print(f"Sum of each row: {row_sums}\n")


# ==========================================
# 2. Pandas: Find the mean of a specific column
# ==========================================
print("--- Q2: Pandas Mean ---")
data = {'Age': [25, 30, 35, 40, 22], 'Salary': [50000, 60000, 75000, 90000, 45000]}
df = pd.DataFrame(data)
salary_mean = df['Salary'].mean()
print(f"DataFrame:\n{df}")
print(f"Mean of 'Salary' column: {salary_mean}\n")


# ==========================================
# 3. Create a scatter plot using Matplotlib
# ==========================================
print("--- Q3: Matplotlib Scatter Plot ---")
x_scatter = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y_scatter = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78]

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='blue')
plt.title("Q3: Matplotlib Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
print("(Scatter plot displayed)\n")


# ==========================================
# 4. Correlation matrix using Seaborn & Heatmap
# ==========================================
print("--- Q4: Seaborn Heatmap ---")
# Using the dataframe from Q2
corr_matrix = df.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Q4: Correlation Matrix Heatmap")
plt.show()
print("(Heatmap displayed)\n")


# ==========================================
# 5. Generate a bar plot using Plotly
# ==========================================
print("--- Q5: Plotly Bar Plot ---")
# Creating simple data for Plotly
df_plotly = pd.DataFrame({
    'Fruit': ['Apples', 'Oranges', 'Bananas'],
    'Amount': [10, 15, 7]
})
fig = px.bar(df_plotly, x='Fruit', y='Amount', title="Q5: Plotly Bar Plot")
fig.show() # This will open in a browser or notebook renderer
print("(Plotly bar chart generated)\n")


# ==========================================
# 6. Add a new column based on an existing column
# ==========================================
print("--- Q6: New DataFrame Column ---")
df['Bonus'] = df['Salary'] * 0.10  # 10% bonus
print(df)
print("\n")


# ==========================================
# 7. Element-wise multiplication of two NumPy arrays
# ==========================================
print("--- Q7: NumPy Element-wise Multiplication ---")
arr_a = np.array([1, 2, 3])
arr_b = np.array([4, 5, 6])
multiplication_result = arr_a * arr_b
print(f"Array A: {arr_a}")
print(f"Array B: {arr_b}")
print(f"Result (A * B): {multiplication_result}\n")


# ==========================================
# 8. Create a line plot with multiple lines using Matplotlib
# ==========================================
print("--- Q8: Matplotlib Multi-line Plot ---")
x_line = np.linspace(0, 10, 100)
y1 = np.sin(x_line)
y2 = np.cos(x_line)

plt.figure(figsize=(8, 4))
plt.plot(x_line, y1, label='Sine Wave', color='r')
plt.plot(x_line, y2, label='Cosine Wave', color='g')
plt.title("Q8: Multiple Lines Plot")
plt.legend()
plt.show()
print("(Line plot displayed)\n")


# ==========================================
# 9. Filter rows where a column value is greater than a threshold
# ==========================================
print("--- Q9: Pandas Filtering ---")
# Filtering employees with salary > 55000 from df created in Q2/Q6
high_earners = df[df['Salary'] > 55000]
print(f"Employees with Salary > 55000:\n{high_earners}\n")


# ==========================================
# 10. Create a histogram using Seaborn
# ==========================================
print("--- Q10: Seaborn Histogram ---")
# Generating random data for distribution
dist_data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(6, 4))
sns.histplot(dist_data, kde=True, color='purple')
plt.title("Q10: Distribution Histogram")
plt.show()
print("(Histogram displayed)\n")


# ==========================================
# 11. Perform matrix multiplication using NumPy
# ==========================================
print("--- Q11: Matrix Multiplication ---")
mat_a = np.array([[1, 2], [3, 4]])
mat_b = np.array([[5, 6], [7, 8]])

# Using dot product (or @ operator)
mat_prod = np.dot(mat_a, mat_b)
print(f"Matrix A:\n{mat_a}")
print(f"Matrix B:\n{mat_b}")
print(f"Product:\n{mat_prod}\n")


# ==========================================
# 12. Load a CSV file and display first 5 rows
# ==========================================
print("--- Q12: Load CSV ---")
# Creating a dummy CSV file first so the script runs
dummy_csv_name = 'pwskills_dummy_data.csv'
df.to_csv(dummy_csv_name, index=False)

# Reading the CSV back
df_loaded = pd.read_csv(dummy_csv_name)
print(f"First 5 rows of loaded CSV:\n{df_loaded.head()}\n")

# Cleanup: Removing the dummy file to keep your folder clean
if os.path.exists(dummy_csv_name):
    os.remove(dummy_csv_name)


# ==========================================
# 13. Create a 3D scatter plot using Plotly
# ==========================================
print("--- Q13: Plotly 3D Scatter ---")
# Using the 'iris' dataset built into plotly for a good 3D example
df_iris = px.data.iris()
fig_3d = px.scatter_3d(df_iris, x='sepal_length', y='sepal_width', z='petal_width',
                       color='species', title="Q13: 3D Scatter Plot")
fig_3d.show()
print("(3D Scatter plot generated)")

print("\n--- ALL TASKS COMPLETED ---")
```
